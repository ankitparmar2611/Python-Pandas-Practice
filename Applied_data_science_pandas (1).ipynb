### Operations and Transformations in Pandas Series__
Operations and transformations in Pandas Series are crucial for modifying, enhancing, and cleaning data effectively.

They provide flexibility to adapt data to specific analyses or visualizations, preparing it for meaningful insights and ensuring data quality.

```python
# Here we can add 2 series with each other.
data = [1,2,3,4,5,6]
new_data = [10,12,13,14,15,16]
series = pd.Series(data)

# Element-wise addition
result_series = series + new_data
print(f"Add new_data + data : \n{result_series}")

# Apply function to each element.

squared_series = series.apply(lambda x:x**2)
print(f"Square of Series : \n{squared_series}")

s = pd.Series([12,13,14,15,16])
new_s = s.apply(lambda x: "Pass" if x >= 14 else "Fail")
print(f"Greater or equal to 14 : \n{new_s}")

# Map values using a dictionary
mapped_series = series.map({1: 'one', 2: 'two', 3: 'three',4:'Four',5:'Five',6:'Six'})
print(f"Mapped series with values : \n{mapped_series}")

# Sort the Series by values
sorted_series = series.sort_values()
print(f"Sorted : \n{sorted_series}")

# Check for missing values
missing_values = series.isnull()
print(missing_values)
```

### Querying a Series__
Selecting and filtering data based on specific conditions is an essential aspect of querying a Pandas Series.

The following examples illustrate common querying operations that can be applied to a Pandas Series:

```python
data = {"a":10,"b":20,"c":30,"d":40,"e":50}

dict_series = pd.Series(data)
print(f"Dictionary : \n{dict_series}")

# Select elements greater then 30.
greater_than_30 = dict_series[dict_series > 30]
print(f"Numbers greater then 30 : \n{greater_than_30}")

# Equals to 20.
equals_to_20 = dict_series[dict_series == 20]
print(f"Numbers equal to 20 : \n{equals_to_20}")

# Select elements not equals to 40.

not_equal_40 = dict_series[dict_series != 40]
print(f"Numbers not equals to 40 : \n{not_equal_40}")

# Selected elements based on multiple conditions.

multiple_conditions = dict_series[(dict_series > 10) & (dict_series < 40)]
print(f"Numbers greater the 10 less and then 40 : \n{multiple_conditions}")

# Checks value is in list or not

check = dict_series[dict_series.isin([20,40,60])]
print(f"Check if numbers exits or not : \n{check}")

# Select elements using string methods (if applicable)
string_series = pd.Series(["Apple","Mango","Banana","Cherry","Blue Berry"])

str_find = string_series[string_series.str.startswith("B")]
print(f"Find string based on starting word : \n{str_find}")

# Query based on index labels
selected_by_index_labels = series_index.loc[['a', 'c', 'e']]
print(f"Get values with index : \n{selected_by_index_labels}")

# Query based on numeric position
selected_by_numeric_position = series_index.iloc[1:4]
print(f"Get values based on numeric position : \n{selected_by_numeric_position}")
```

### Correlation Analysis
The corr() function generates a correlation matrix, indicating how variables relate to each other.

Values closer to 1 or -1 imply a stronger correlation, while values near 0 suggest a weaker correlation.

```python
import pandas as pd

# Create a sample DataFrame with numeric columns
# Each key becomes a column, and each list becomes column values
data = {
    'Numeric_column1': [5, 15, 8],
    'Numeric_column2': [10, 20, 30],
    'Numeric_column3': [100, 200, 300]
}

# Convert dictionary into DataFrame (tabular format)
df = pd.DataFrame(data)

# Compute correlation matrix
# .corr() calculates correlation between all numeric columns
# Correlation tells how strongly two columns are related
# Value range:
# +1  -> Perfect positive correlation (both increase together)
#  0  -> No correlation (no relationship)
# -1  -> Perfect negative correlation (one increases, other decreases)

correlation_matrix = df.corr()

# Example explanation of correlation in this dataset:
# Numeric_column2 and Numeric_column3:
# As column2 increases (10 → 20 → 30),
# column3 also increases (100 → 200 → 300)
# → This gives strong positive correlation (~1)

# Numeric_column1 and Numeric_column2:
# Values are not increasing in a consistent pattern together
# (5 → 15 → 8 vs 10 → 20 → 30)
# → So correlation is weaker (not close to 1)

# Numeric_column1 and Numeric_column3:
# Same reason → not perfectly aligned → weaker correlation

# Print correlation matrix
print("Correlation matrix:\n", correlation_matrix)
```

```python
import pandas as pd

# -------------------------------
# Step 1: Create DataFrame
# -------------------------------
data = {
    "Department": ["HR", "IT", "HR", "FINANCE", "IT", "FINANCE"],
    "Employee": ["A", "B", "C", "D", "E", "F"],
    "Salary": [45000, 50000, 30000, 40000, 60000, 30000]
}

df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:\n", df)


# -------------------------------
# Step 2: Average Salary per Department
# -------------------------------
# Group by Department and calculate mean salary
dept_avg_salary = df.groupby("Department")["Salary"].mean()

# Extract specific department averages
hr_dept = dept_avg_salary["HR"]
it_dept = dept_avg_salary["IT"]
finance_dept = dept_avg_salary["FINANCE"]

print("\nDepartment-wise Average Salary:\n", dept_avg_salary)
print(f"\nAverage salary of HR department: {hr_dept}")
print(f"Average salary of IT department: {it_dept}")
print(f"Average salary of Finance department: {finance_dept}")


# -------------------------------
# Step 3: Employee Count per Department
# -------------------------------
# Count number of employees in each department
employee_counts = df.groupby("Department")["Employee"].count()

print("\nDepartment-wise Employee Counts:\n", employee_counts)

# Total employees in company
total_employee = employee_counts.sum()
print(f"\nTotal Employees: {total_employee}")

# Extract individual department counts
employee_hr = employee_counts["HR"]
employee_it = employee_counts["IT"]
employee_finance = employee_counts["FINANCE"]

print("\nDepartment-wise Separate Employee Counts:")
print(f"HR: {employee_hr}")
print(f"IT: {employee_it}")
print(f"Finance: {employee_finance}")


# -------------------------------
# Step 4: Total Salary per Department
# -------------------------------
# Sum of salaries for each department
employee_salary = df.groupby("Department")["Salary"].sum()

print("\nTotal Salaries Department-wise:\n", employee_salary)

# Total salary expense of company
total_sal = employee_salary.sum()
print(f"\nTotal Salary Expense: {total_sal}")


# -------------------------------
# Step 5: Multiple Aggregations
# -------------------------------
# Apply multiple aggregate functions on Salary
employee_salary_details = df.groupby("Department")["Salary"].agg(["mean", "min", "max"])

print("\nEmployee Salary Details (mean, min, max):\n", employee_salary_details)
```

```python
import pandas as pd

# --------------------------------------
# Step 1: Create DataFrame
# --------------------------------------
# Dictionary where:
# - Keys = column names
# - Values = column data
data = {
    "Department": ["HR", "IT", "HR", "FINANCE", "IT", "FINANCE"],
    "Employee": ["A", "B", "C", "D", "E", "F"],
    "Emp_id": [12, 13, 14, 15, 16, 17],
    "Salary": [45000, 50000, 30000, 40000, 60000, 30000]
}

# Convert dictionary into DataFrame (table format)
df = pd.DataFrame(data)

# --------------------------------------
# Step 2: Iterating over groups
# --------------------------------------
# groupby("Department") will:
# - Split the DataFrame into groups based on Department
# - Each group contains FULL rows (not just Department column)

# Loop will run once for each unique department (HR, IT, FINANCE)
for dept, group in df.groupby("Department"):

    # 'dept' = current group name (HR / IT / FINANCE)
    print("Department:", dept)

    # 'group' = mini DataFrame containing ALL columns
    # (Employee, Emp_id, Salary) for that department only
    # Example:
    # If dept = "HR", then group will contain only HR rows
    print(group)

    # Important Understanding:
    # - Loop is NOT running on only Department column
    # - It is running on grouped DataFrames
    # - Each 'group' contains complete rows for that department

    # Internally it works like:
    # HR → all rows where Department = HR
    # IT → all rows where Department = IT
    # FINANCE → all rows where Department = FINANCE

    # So Employee, Emp_id, Salary come automatically
    # because group contains full filtered data
```

```python
# --------------------------------------
# Get specific group using groupby
# --------------------------------------

# groupby("Department") creates groups based on Department
# .get_group("HR") selects only the "HR" group

hr_data = df.groupby("Department").get_group("HR")

# 'hr_data' now contains only rows where Department = HR
# It returns a mini DataFrame with all columns (Employee, Emp_id, Salary)

print(hr_data)

# Example Output:
#   Department Employee  Emp_id  Salary
# 0         HR        A      12   45000
# 2         HR        C      14   30000

# Key Understanding:
# - groupby() splits data into groups
# - get_group("HR") picks only HR group
# - Output is a filtered DataFrame (not just one column)
```

### House Prices Data sets

```python
import pandas as pd

df = pd.read_csv("datasets/HousePrices.csv")
print(df.head())
```

```python
# Data Cleaning

print("Data Information : \n")
print(f"{df.info()}\n")

print(f"Data Shape : \n{df.shape}")

# Remove nulls
df.dropna(inplace=True)

# Remove invalid values
df = df[df['price'] > 0]
df = df[df['bedrooms'] > 0]
df = df[df['bathrooms'] > 0]

# Fix renovated year
df.loc[df['yr_renovated'] == 0, 'yr_renovated'] = df['yr_built']

# Convert date
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year


# Remove outliers
df = df[df['price'] < df['price'].quantile(0.99)]

# Reset index
df.reset_index(drop=True, inplace=True)

print("Data cleaned successfully ✅")
print(f"{df.info()}\n")
df.isnull().sum()
df.describe()




    
```
